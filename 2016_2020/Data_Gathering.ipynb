```python
import pandas as pd


def filter_general_rows(df):
    # Ensure the first column is treated as strings
    df.iloc[:, 0] = df.iloc[:, 0].astype(str)
    # Filter rows where the first column contains the word "General"
    filtered_df = df[df.iloc[:, 0].str.contains(r'\bGeneral\b', case=False, na=False)]
    return filtered_df


filtered_df = filter_general_rows(df)
filtered_df
```
